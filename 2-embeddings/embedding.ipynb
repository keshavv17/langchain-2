## Embedding techniques
converting text into vectors

```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY")
```

```python
from langchain_openai import OpenAIEmbeddings
embeddings = OpenAIEmbeddings(model="text-embedding-3-large")
embeddings
```

```python
text = "this is a tutorial on openai embeddings"
queryresult = embeddings.embed_query(text)
queryresult, len(queryresult)
```

```python
from langchain_community.document_loaders import TextLoader
loader = TextLoader('speech.txt')
docs = loader.load()
docs
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size= 500, chunk_overlap = 50)
final_doc = text_splitter.split_documents(docs)
final_doc
```

## Vector embedding and vector store DB

```python
from langchain_community.vectorstores import Chroma

db = Chroma.from_documents(final_doc, embeddings)
db
```

```python
### retrieve the results from chromadb store
query = "Success is not just about reaching a destination"
retrieved_res = db.similarity_search(query)
print(retrieved_res)
```
